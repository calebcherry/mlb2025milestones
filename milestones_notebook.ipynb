### Blue Jays: Vlad for average and power
Tony Fernandez has the most seasons with at least a .310 BA. Vlad Jr has done it twice so far. Will 2025 be Year 3?
Also, since 2021 Vlad Jr has the second most batted balls with above 115 mph exit velo. Stanton has the most at 95.

```python
# Create a function that takes df_batting_2025 and returns a True/False value if Vlad's batting average is currently above .310 and also returns the current batting average.
def is_vlad_batting_above_310(df, id_fg):
    # Get Vlad's batting average
    vlad_batting_avg = df.loc[df['IDfg'] == id_fg, 'AVG'].values[0]
    
    # Check if batting average is above .310
    if vlad_batting_avg > .310:
        return True, vlad_batting_avg
    else:
        return False, vlad_batting_avg
    
# TODO: Research what counts as a batted ball to also track number of batted balls over 115 mph by Vlad Jr.
```
